```python
xlsx = {
    'file'  : '/mnt/d/autocls_e3/trainref/pixels_exp32.xlsx',
    'sheet' : 'cosc2021',
    'pxid'  : 'rootid',
    'classcol' : 'expclass'
}

shp = {
    'file' : '/mnt/d/autocls_e3/classref/refexp32_vec.shp',
    'pxid' : 'pxid',
    'classcol' : 'nameval'
}

outres = '/mnt/d/autocls_e3/trainref/pixels_exp32a.xlsx'
```

```python
import copy
from glass.rd import tbl_to_obj
from glass.rd.shp import shp_to_obj
from glass.wt import obj_to_tbl
```

```python
tdf = tbl_to_obj(xlsx['file'], sheet=xlsx['sheet'])
sdf = shp_to_obj(shp['file'])

ocols = copy.deepcopy(tdf.columns.values)

tdf.drop([xlsx['classcol']], axis=1, inplace=True)

tdf = tdf.merge(
    sdf, how='left',
    left_on=xlsx['pxid'], right_on=shp['pxid']
)

tdf.rename(columns={shp['classcol'] : xlsx['classcol']}, inplace=True)
tdf.drop([c for c in tdf.columns.values if c not in ocols], axis=1, inplace=True)
```

```python
obj_to_tbl(tdf, outres, sheetsName=xlsx['sheet'])
```
